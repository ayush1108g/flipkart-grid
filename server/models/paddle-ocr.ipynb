```python
from paddleocr import PaddleOCR,draw_ocr
ocr = PaddleOCR(use_angle_cls=True, lang='en')
```

```python
img_path = './1.jpg'
# img_path = '/kaggle/input/image1/11.png'
result = ocr.ocr(img_path, cls=True)


for idx in range(len(result)):
    res = result[idx]
    for line in res:
        print(line)

extracted_text = ""
for idx in range(len(result)):
    res = result[idx]
    for line in res:
        extracted_text += line[1][0] + " "

print("Extracted Text: ", extracted_text)

# draw result
from PIL import Image
result = result[0]
image = Image.open(img_path).convert('RGB')
boxes = [line[0] for line in result]
txts = [line[1][0] for line in result]
scores = [line[1][1] for line in result]
im_show = draw_ocr(image, boxes, txts, scores,  font_path="/kaggle/input/dataset2/OCR/Times/times.ttf")
im_show = Image.fromarray(im_show)
im_show.save('/kaggle/working/result.jpg')
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load and display an image
img = mpimg.imread('/kaggle/working/result.jpg')
plt.imshow(img)
plt.axis('off')  # Optional: Turn off axis labels
plt.show()
```
